```python
import gensim
```

```python
from gensim.models import Word2Vec
```

```python
import gensim.downloader as api
```

```python
info = api.info()
corpus = api.load('text8')
#corpusFake =api.load('fake-news')
model = Word2Vec(corpus)  # train a model from the corpus
```

```python
model.save("text8_trained")
```

```python
model.wv.save_word2vec_format("text8_trained_format")
```
